## Задание 2

О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
Если да, найдите ее.

```python
D = 0.2
a = 0.5
```

Из формулы для дисперсии раномерного распределения можно выразить правую границу интервала: <br>
$b = \sqrt{12D}+a$

```python
b = (12 * D)**0.5 + a
print(round(b,3))
```

Вычислить её среднее значение можно по формуле: $M(B) = \frac{a+b}{2}$

```python
M = (a + b) / 2
print(round(M, 2))
```

**Ответ:** Правая граница для случайной величины $B$ равна 2.049, а её среднее значение - 1.27
